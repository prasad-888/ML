```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = np.array([
    [300, 50], #Alice
    [400, 65], #Bob
    [800,120], #Charlie
    [700, 110], #Diana
    [200, 30], #EVe
    [150, 25], #Frank
])
```

```python
# k-Means Clustering 
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```

```python
# Cluster Centers and Labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
print(centers)
```

```python
print(labels)
# label as in label of the data
```

```python
#PLot
plt.scatter(X[:, 0], X[:,1], c=labels, cmap='cividis', marker='o', label='Data Points')
plt.scatter(centers[:,0],centers[:,1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering of Spending Habits')
plt.xlabel("Monthly Spending ($)")
plt.ylabel('Number of Purchases')
plt.legend()
plt.show()
```

### The Elbow Method 📈

The **Elbow Method** is a popular technique used to determine the **optimal number of clusters (K)** for a K-Means clustering algorithm. It's a simple visual tool that helps you select a good value for **K** by finding the "elbow" in the plot, which is the point of diminishing returns.

***

### How It Works 💡

The Elbow Method works by running the K-Means algorithm for a range of **K** values and calculating a metric called **inertia** for each run.

* **K-Means Algorithm:** K-Means is an iterative clustering algorithm that partitions a dataset into **K** distinct, non-overlapping subgroups (clusters).
* **Inertia:** Inertia is the **sum of squared distances** of samples to their closest cluster center. It's a measure of how **compact** and **well-separated** the clusters are. The lower the inertia, the better the clusters.

***

### The Process ✍️

The process involves these steps:

1.  **Run K-Means for a range of K values:** You start by running the K-Means algorithm multiple times, each time with a different number of clusters (e.g., K = 1, 2, 3, 4, 5, and so on).
2.  **Calculate Inertia:** For each value of **K**, you calculate the **inertia**.
3.  **Plot the Elbow Curve:** You then plot the **K values** on the x-axis and the **inertia** on the y-axis. The resulting graph is called the **Elbow Curve**.
4.  **Find the "Elbow" Point:** You look for the point on the curve where the rate of decrease in inertia **slows down significantly**. This point forms an "elbow" shape on the plot. The **K value** at this elbow is considered the optimal number of clusters.



***

### `KMeans.inertia_` Explained 📊

`KMeans.inertia_` is an attribute in scikit-learn's `KMeans` class that represents the **within-cluster sum of squares (WCSS)**. It measures how compact or cohesive the clusters are. A lower inertia value indicates that the data points within a cluster are closer to their respective cluster's centroid, meaning the clusters are tighter and more well-defined.

---

### How Inertia is Calculated ➗

The inertia for a K-Means model is calculated by:

1.  For each data point, determining its distance to its assigned cluster's centroid.
2.  Squaring that distance.
3.  Summing all these squared distances for every data point across all clusters.

The formula can be expressed as:

$$Inertia = \sum_{i=1}^{n} (x_i - \mu_{c_i})^2$$

where:
* $x_i$ is a data point.
* $\mu_{c_i}$ is the centroid of the cluster to which data point $x_i$ belongs.
* The summation is over all $n$ data points in the dataset.

Because of its nature, inertia naturally decreases as the number of clusters (K) increases. With more clusters, each data point can be assigned to a centroid that's closer to it, thus reducing the total sum of squared distances. The goal is to find a balance between having a low inertia and a reasonable number of clusters. This is precisely what the **Elbow Method** helps visualize.

***

### Why We Declare `inertia = []` 📋

You declare `inertia = []` because it's a **list** that will be used to store the `inertia_` value calculated for each K-Means model. The code iterates through a range of `k_values` (the number of clusters), and for each `k`, it trains a new K-Means model.

Here's a breakdown of why this is necessary:

* **Initialization:** The `inertia = []` line creates an **empty list** before the loop begins. This list will be the container for all the inertia scores.
* **Looping and Calculation:** Inside the `for` loop, for each value of `k`, the code does the following:
    1.  `kmeans = KMeans(n_clusters=k, random_state=42)`: It creates a K-Means model with a specific number of clusters, `k`.
    2.  `kmeans.fit(X)`: It trains the model on the dataset `X`.
    3.  `inertia.append(kmeans.inertia_)`: After the model is trained, it accesses the `inertia_` attribute of that specific model instance and **appends** (adds) it to the `inertia` list.
* **Storing Results:** This process ensures that after the loop finishes, the `inertia` list contains a series of values—one for each `k` tested. This list is then used to create the **Elbow Curve plot**, where `k_values` are plotted against their corresponding `inertia` scores.

```python
# Elbow Method 

inertia = []
k_values = range(1, 6)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
```

```python
# Plot Elbow  Curve

plt.plot(k_values, inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of CLusters (K)")
plt.ylabel("Inertia")
plt.show()
```
