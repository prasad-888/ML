IN PCA or Principle COmp-onent Analysis

__high dimensional data -> transform -> new axes called principal components__

__How PCA Works__

1. __Standardization__: Every feature is scaled so they contribute equally to the analysis.
2. __Covariance Matrix__: Relationships between features are analyzed.
3. __Eigenvectors and Eigenvalues__: Principal components are identified based on variance.
4. __Projection__: Data is transformed into the reduced-dimensional space.

### **Difference Between `print(iris)` and `print(iris.data)`**

When you load the Iris dataset in scikit-learn using `iris = load_iris()`, there's a key difference between printing the entire object and printing its data attribute.

---

### **`print(iris)`**

Printing `iris` displays a **Bunch** object. This is a dictionary-like container that holds everything related to the dataset. It's not just the raw numbers but also metadata that makes the dataset understandable and easy to work with.

The `Bunch` object contains:
* **`data`**: The numerical feature data as a NumPy array.
* **`target`**: The class labels (0, 1, 2) for each data point.
* **`target_names`**: The human-readable names for the target labels (e.g., 'setosa').
* **`feature_names`**: The names of the features (e.g., 'sepal length (cm)').
* **`DESCR`**: A full description of the dataset, including its source and characteristics.

Essentially, `print(iris)` gives you a complete overview of the dataset's structure, which is helpful for understanding its components before you start building a model.

---

### **`print(iris.data)`**

Printing `iris.data` accesses only the **feature data** from the `Bunch` object. The output is a **NumPy array** containing the raw numerical values of the four features for each of the 150 samples.

This array has a shape of (150, 4), which means it has 150 rows (one for each flower) and 4 columns (one for each feature). This is the part of the dataset you would directly use for training a machine learning model.

In short, `print(iris)` shows you the entire dataset package, while `print(iris.data)` isolates and displays just the numerical feature matrix you'll be feeding into your algorithm.

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
target = iris.target
```

# What are Eigenvalues and Eigenvectors? ðŸ§
Eigenvectors are special vectors that don't change their direction when a linear transformation is applied to them. They can, however, be scaled. Think of them as the principal axes of a transformation. They point in the most important directions of variance in your data.



Eigenvalues are the scalar values that correspond to each eigenvector. They tell you the magnitude of the scaling that occurs in the direction of the eigenvector. A larger eigenvalue means a larger scaling, which in the context of PCA, signifies a greater amount of variance in that direction.



In essence, an eigenvector represents a direction, and its corresponding eigenvalue represents the strength or importance of that direction.

# How PCA Uses Them ðŸ“Š
Principal Component Analysis (PCA) is a dimensionality reduction technique that finds new, uncorrelated features called principal components. PCA uses eigenvectors and eigenvalues to identify these components.


1. Calculate the Covariance Matrix: First, PCA calculates the covariance matrix of the data. This matrix summarizes how each feature in the dataset varies with every other feature.

2. Find Eigenvectors and Eigenvalues: PCA then computes the eigenvectors and eigenvalues of this covariance matrix.

3. Identify Principal Components: The eigenvectors of the covariance matrix are the principal components. They form a new set of orthogonal axes. The eigenvector with the largest eigenvalue is the first principal component, as it captures the most variance in the data. The eigenvector with the second largest eigenvalue is the second principal component, and so on.



4. Select Components: You can then choose to keep only the top k principal components (the ones with the largest eigenvalues) and discard the rest. This reduces the dimensionality of your dataset while retaining as much of the original data's variance as possible.


The eigenvalues help you decide how many principal components to keep. By plotting the eigenvalues in a scree plot, you can visually determine the "elbow" where the eigenvalues start to drop off dramatically. This point suggests an appropriate number of components to retain for your analysis.



In summary, eigenvectors define the new axes of your data, and eigenvalues tell you how important those axes are in explaining the variance in your data.



# n_components is a parameter in scikit-learn's PCA class that specifies the number of principal components to keep after dimensionality reduction.

# How It Works âš™ï¸
The PCA algorithm works by identifying the principal components, which are the directions of maximum variance in your data. By default, it will find as many principal components as there are features in your original dataset. However, the power of PCA lies in reducing dimensionality by keeping only the most important components.

The n_components parameter allows you to control this process:

* n_components=2: This is a very common setting for visualization. It instructs PCA to find the top two principal components (the ones that explain the most variance) and project the data onto these two new axes. This reduces the data from its original number of dimensions down to just two, making it easy to plot on a 2D graph.

* Other Values: You can set n_components to any integer from 1 up to the number of features in your dataset. For example, n_components=3 would reduce the data to a 3D space. You can also set it to a floating-point number between 0 and 1, which represents the percentage of variance you want to retain. For example, PCA(n_components=0.95) would select the minimum number of principal components needed to explain 95% of the total variance.

In the code snippet pca = PCA(n_components=2), you are creating a PCA object that is configured to reduce your dataset to two dimensions. This is often the first step in a workflow for visualizing high-dimensional data, as it allows you to see the relationships and clusters in your data in a simplified form.

```python
# Apply PCA

pca =PCA(n_components=2)
reduced_data = pca.fit_transform(data)
```

```python
reduced_data
```

# pca.explained_variance_ratio_ is an attribute in scikit-learn's PCA class that shows the percentage of variance explained by each of the selected principal components. It's a key metric for understanding how much information each component retains from the original dataset.

# How It's Calculated and Used ðŸ“ˆ
The explained_variance_ratio_ is calculated by dividing the eigenvalue of each principal component by the sum of all eigenvalues. The result is a list or array where each value corresponds to a principal component and represents the proportion of the total variance in the dataset that is captured by that specific component.

1. Eigenvalues: As discussed previously, eigenvalues represent the amount of variance along each principal component.

2. Total Variance: The sum of all eigenvalues represents the total variance in the dataset.

3. Proportion: The explained_variance_ratio_ for a single component is its eigenvalue divided by the total variance.

For example, if the first principal component has an explained_variance_ratio_ of 0.85, it means that this single component accounts for 85% of the total variance in the original data.

This attribute is crucial for:

* Determining the number of components: By looking at the explained_variance_ratio_, you can decide how many principal components are needed to explain a significant portion of the total variance. A common practice is to choose the number of components that collectively account for 95% or 99% of the variance.

* Creating a Scree Plot: The values from explained_variance_ratio_ are often used to create a scree plot, which plots the variance explained by each component. This plot helps to visually identify the "elbow" point, where adding more components provides diminishing returns in terms of explained variance.



```python
# Plot the reduced data

plt.scatter(reduced_data[:, 0], reduced_data[:,1], c=target, cmap='cividis', edgecolor='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Reduction of IRIS')

plt.colorbar(label='Class')
plt.show()

# Explained variance ratios
print("Explained Variance Ratios:", pca.explained_variance_ratio_)
```
