# t- Distributed Stochastic Neighbor Embedding (t-SNE)

# technique -> non-linear -> preserve local structure of data  -> ensuring similar points stay close while dissimilar ones are far apart

# How t-SNE Works
1. __Pairwise Similarities__: Calculates probabilities of similarity in high-dimensional space.
2. __Low-Dimensional Mapping__: Maps points to lower dimensions while preserving these similarities.
3. __Optimization__: Minimizes the difference between high and low-dimensional similarities.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(data= iris.data, columns = iris.feature_names)
target = iris.target
```

```python
data
```

# A manifold in machine learning is a low-dimensional subspace that a dataset resides in, even if the data appears to be in a high-dimensional space. Think of it like a crumpled-up piece of paper. The paper itself is a 2D object, but when crumpled, it exists in a 3D space. The flat, 2D surface is the manifold. Manifold learning algorithms aim to "un-crumple" this data to reveal its true, underlying structure.

Why Manifolds are Important in ML üåê
Many datasets, especially high-dimensional ones like images, text, or genomic data, are not uniformly spread out in their entire high-dimensional space. Instead, they cluster around a lower-dimensional structure‚Äîthe manifold.

Manifold learning is a type of non-linear dimensionality reduction. Unlike linear methods like PCA, which project data onto a flat hyperplane, manifold learning algorithms can discover and preserve the non-linear relationships and local structure of the data.

Common manifold learning techniques include:

t-SNE (t-Distributed Stochastic Neighbor Embedding): Excellent for visualizing high-dimensional data by placing similar data points near each other in a low-dimensional map.

Isomap (Isometric Mapping): Works by preserving the geodesic distances (the shortest path along the curved manifold) between data points.

LLE (Locally Linear Embedding): Assumes that each data point and its neighbors lie on a local, linear patch of the manifold. It preserves these local relationships to create a lower-dimensional embedding.

Using manifold learning can be beneficial for:

Visualization: It allows you to plot and understand complex, high-dimensional data in 2D or 3D.

Improved Performance: By reducing noise and non-informative dimensions, a model trained on the manifold-reduced data may perform better.

Feature Extraction: The low-dimensional representation can serve as a set of new, more meaningful features for subsequent machine learning tasks.

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=42)
tsne_data = tsne.fit_transform(data)
```

```python
tsne_data
```

# As you see we have reduced the data into columns

```python
#plot

plt.scatter(tsne_data[:,0], tsne_data[:,1], c=target, cmap='viridis', edgecolor='k')
plt.title('t-SNE Reduction of Iris ')
plt.colorbar(label='Class')
plt.show()
```
