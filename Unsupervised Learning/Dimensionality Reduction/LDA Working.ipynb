# LDA or Linear Discriminant Analysis
LDA -> __maximize__ -> __seperation between classes__


# [article](https://sebastianraschka.com/Articles/2014_python_lda.html)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
target = iris.target
```

![](https://sebastianraschka.com/images/blog/2014/linear-discriminant-analysis/lda_1.png)

```python
data
```

```python
target
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# apply LDA
lda = LDA(n_components=2)
lda_data =  lda.fit_transform(data, target)
```

```python
lda_data
```

# as you see we have reduced reduced the data from petal, sepal length width and target to two columns

```python
# PLot the reduced data 

plt.scatter(lda_data[:,0], lda_data[:,1], c=target, cmap= 'viridis', edgecolor='k')
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title("LDA Reduction of IRis")
plt.colorbar(label='Class')
plt.show()
```
