```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\Prasad\Desktop\ml\data_ml\Fish.csv")
```

```python
df.head()
```

```python
y =  df['Species'] ## labels
```

```python
y.head()
```

```python
X = df.drop(['Species'], axis=1) ## features
```

```python
X.head()
```

```python

knn = KNeighborsClassifier(n_neighbors=12,p=1) ## p=1 -> Manhattan

```

```python
knn.fit(X,y)
```

```python
print(knn.score(X,y))
```

```python
sample = np.array([12,56,234,66,123,56])

knn.predict([sample])
```
