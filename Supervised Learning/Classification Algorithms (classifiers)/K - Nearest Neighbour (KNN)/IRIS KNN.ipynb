```python
# importing packages

import numpy as np
import pandas as pd
import os
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## reading the dataset 

iris = pd.read_csv(r"C:\Users\Prasad\Desktop\ml\data_ml\iris.csv")
```

```python
iris.head()
```

```python
X = iris.drop(['class' , 'Unnamed: 0'],axis=1)
```

```python
X ## features
```

```python
y = iris['class']
```

```python
y ## labels or targets
```

```python
## Initializing and training the model

knn = KNeighborsClassifier(n_neighbors=15, metric='manhattan') # n_neighbors - k value and metric - distance formula
## n_neighbors -> K value   metric-> distance formula used
```

```python
knn.fit(X,y) ## initialization of model training
```

```python
## Predicting the class using sample data

sample = np.array([4.5, 3.2, 1.6, 0.5])


## knn object of KNeighborsClassifier
print(knn.predict([sample]))
```

```python
    import warnings
    warnings.filterwarnings("ignore")
```

```python
print(knn.predict([sample]))
```

```python
## accuracy
print(knn.score(X,y)) ## Accuracy of the trained model
```

```python
score_list_for_various_k =[]
```

```python
for i in range(1,101):
    kn= KNeighborsClassifier(n_neighbors=i,metric="manhattan")
    kn.fit(X,y)
    score_list_for_various_k.append(kn.score(X,y))
```

```python
score_list_for_various_k
```

```python
plt.plot(data=score_list_for_various_k, type='line')
plt.show()
```

```python
df = pd.DataFrame(score_list_for_various_k)
```

```python
df
```

```python
df.plot(kind='line', title='Line Plot for K values From 0-100',xlabel='k',ylabel='accuracy')
plt.show()
```
