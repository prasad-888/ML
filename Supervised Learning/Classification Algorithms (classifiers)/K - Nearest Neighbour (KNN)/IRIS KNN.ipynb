```python
# importing packages

import numpy as np
import pandas as pd
import os
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## reading the dataset 

iris = pd.read_csv(r"C:\Users\Prasad\Desktop\ml\data_ml\iris.csv")
```

```python
iris.head()
```

```python
X = iris.drop(['class' , 'Unnamed: 0'],axis=1)
```

```python
X ## features
```

```python
y = iris['class']
```

```python
y ## labels or targets
```

```python
y.value_counts()
```

```python
## Initializing and training the model

knn = KNeighborsClassifier(n_neighbors=15, metric='manhattan') # n_neighbors - k value and metric - distance formula
## n_neighbors -> K value   metric-> distance formula used
```

```python
knn.fit(X,y) ## initialization of model training
```

```python
## Predicting the class using sample data

sample = np.array([4.5, 3.2, 1.6, 0.5])


## knn object of KNeighborsClassifier
print(knn.predict([sample]))
```

```python
## accuracy
print(knn.score(X,y)) ## Accuracy of the trained model
```

```python
score_list_for_various_k =[]
```

```python
for i in range(1,101):
    kn= KNeighborsClassifier(n_neighbors=i,metric="manhattan")
    kn.fit(X,y)
    score_list_for_various_k.append(kn.score(X,y))
```

```python
score_list_for_various_k
```

```python
plt.plot(data=score_list_for_various_k, type='line')
plt.show()
```

```python
from sklearn.model_selection import learning_curve
from sklearn.model_selection import learning_curve, StratifiedKFold
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
train_sizes, train_scores, test_scores = learning_curve(
    knn, X, y,
    train_sizes=np.linspace(0.1, 1.0, 10),
    cv=cv,
    scoring='accuracy',
    shuffle=True,
    random_state=42
)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Training Score')
plt.plot(train_sizes, test_mean, 'o-', color='green', label='Validation Score')

plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.2, color='blue')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.2, color='green')

plt.title('Learning Curve (KNN on Iris Dataset)')
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```
