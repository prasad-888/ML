What does oob_score_ represent?
The oob_score_ gives the accuracy of the model on the samples that were not used in training each individual tree. These samples are effectively used as a cross-validation set to estimate the model's performance.

It provides a way to validate the model's performance without the need for an additional validation set or cross-validation, making it computationally efficient.

Important Notes:
OOB Score Limitations: OOB score is only available when bootstrap=True, which is the default setting for RandomForestClassifier. If you set bootstrap=False, you wonâ€™t get the oob_score.

When to Use OOB: OOB score is a good internal validation score, but for final model evaluation, you should still use a hold-out test set (i.e., X_test and y_test) or cross-validation to ensure the robustness of the model.
