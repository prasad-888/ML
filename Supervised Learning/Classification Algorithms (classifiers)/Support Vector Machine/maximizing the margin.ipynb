### Maximizing the margin

Support Vector Machines (SVMs) provide a better approach than simply drawing a thin line between classes. The key idea is to create a "margin" or buffer zone around the dividing line (hyperplane). This margin extends outward from the line on both sides, stopping at the nearest data points of each class. The following visual example illustrates this concept.

support vector machines

In Support Vector Machines (SVMs), the best dividing line (hyperplane) is the one that creates the widest possible margin. This line is chosen as the optimal model for separating the data. Because SVMs focus on maximizing this margin, they are known as maximum margin estimators.

Support vectors are the crucial data points nearest to the dividing line (hyperplane). They directly determine the orientation of the line. By focusing on these support vectors, the SVM maximizes the margin (the gap between the classes). If you were to remove these support vectors, the hyperplane would shift. In essence, these are the key data points that define the SVM model.
