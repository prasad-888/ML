__Why and how cross validation is used?__
-> As we know we divide the dataset into training set and testing set
The __purpose__ of cross-validation is to provide a better estimate of a model's ability to perform on unseen data. It provides an unbiased estimate of the generalization error, especially in the case of limited data.

There are many reasons we may want to do this:

To have a clearer measure of how our model performs.
To tune hyperparameters.
To make model selections.
The intuition behind cross-validation is simple - rather than training our models on one training set we train our model on multiple subsets of data.
__Steps:__
1. Split data into portions.
2. Train our model on a subset of the portions.
3. Test our model on the remaining subsets of the data.
4. Repeat steps 2-3 until the model has been trained and tested on the entire dataset.
5. Average the model performance across all iterations of testing to get the total model performance.
   

__Why Good fit algorithms is prefered?__
-> In machine learning, a "good fit" refers to a model that strikes a balance between accurately predicting outcomes on the training data and generalizing well to unseen data. It avoids both overfitting (memorizing the training data too well) and underfitting (failing to capture underlying patterns). A good fit is crucial because it ensures the model will perform effectively in real-world scenarios. 
Here's why a good fit is desirable:
Accurate Predictions:
A good fit means the model's predictions are close to the actual values, leading to reliable results. 
Generalization:
It allows the model to make accurate predictions on new, unseen data, not just the data it was trained on. 
Avoids Overfitting and Underfitting:
Overfitting leads to poor generalization, while underfitting means the model is not learning the data effectively. 
Cost-Effective:
A model that fits well can save time and resources by requiring less retraining and fewer adjustments. 
Reliable Insights:
A good fit allows for more reliable insights from the data, leading to better decision-making. 
In essence, the goal of many machine learning algorithms is to find this "sweet spot" of a good fit, where the model is neither too complex nor too simple, but just right for the task at hand. 

__Define Overfitting & Underfitting.__
->Overfitting occurs when a machine learning model learns the training data too well, including the noise and random fluctuations, leading to poor performance on new, unseen data. Essentially, it becomes too specific to the training set and doesn't generalize well.

__What is model complexity? What happens when model complexity increases?__
What is Model Complexity?
Structure: The architecture of the model, such as the number of layers in a neural network or the degree of a polynomial in regression. 
Number of Parameters: More parameters generally mean a more complex model. 
Capacity: The ability of the model to capture complex patterns in the data. 
What Happens When Model Complexity Increases?
Improved Training Performance:
More complex models can fit the training data more accurately, potentially leading to lower training error. 
Overfitting:
If the model becomes too complex, it might start memorizing the training data, including noise and outliers. This leads to poor generalization to new, unseen data. 
Increased Risk of Overfitting:
High complexity can lead to a situation where the model performs exceptionally well on the training data but poorly on new data. 
Increased Computational Cost:
Complex models often require more computational resources for training and inference. 
Potential for Underfitting:
While complex models can overfit, overly simple models can underfit, failing to capture the underlying patterns in the data. 
Bias-Variance Trade-off:
Increasing complexity generally reduces bias (the error from overly simplifying assumptions) but increases variance (the sensitivity to fluctuations in the training data). 
In essence, finding the right level of model complexity is crucial. You want a model that is complex enough to capture the underlying patterns in your data without becoming so complex that it overfits the training data. 

__In SVM, how does values of C & kernel affect model complexity?__
In Support Vector Machines (SVMs), the C parameter and the kernel significantly influence model complexity. A higher C value leads to a more complex model with a smaller margin and potentially lower training error, while a lower C value results in a simpler model with a larger margin and potentially higher training error. The choice of kernel (linear, polynomial, RBF, etc.) also dramatically affects complexity, with RBF and polynomial kernels allowing for more complex, non-linear decision boundaries compared to the linear kernel. 
Support Vector Machines (SVM) Made Simple & How To Tutorial
Here's a more detailed breakdown:
C Parameter:
High C:
Indicates a strong penalty for misclassifications. The SVM aims to fit the training data as closely as possible, potentially leading to a complex model that overfits the training data, especially with non-linear kernels. This results in a smaller margin and more complex decision boundary. 
Low C:
Implies a weaker penalty for misclassifications. The SVM will prioritize a larger margin, even if it means misclassifying some training points. This leads to a simpler model with a larger margin and a less complex decision boundary. 
Kernel:
Linear Kernel:
The simplest kernel, producing a linear decision boundary. It's best suited for linearly separable data and results in a simpler model. 
Polynomial Kernel:
Allows for non-linear decision boundaries by introducing polynomial terms. The degree of the polynomial (controlled by another hyperparameter) affects the complexity. 
Radial Basis Function (RBF) Kernel:
One of the most popular kernels, capable of creating highly complex, non-linear decision boundaries. It uses a 'gamma' parameter to control the influence of each data point, with higher gamma values leading to more complex boundaries. 
Sigmoid Kernel:
Another non-linear kernel that can create complex decision boundaries. It's less commonly used than RBF and polynomial kernels. 
In essence:
Choosing a high C value with a non-linear kernel (RBF or polynomial) can lead to a highly complex model that may overfit the training data. 
Choosing a low C value with a linear kernel can lead to a simpler model that may underfit the data. 
Finding the right balance between C and the kernel (and other hyperparameters like gamma in RBF) is crucial for building a robust SVM model that generalizes well to unseen data.
