### Soft Margins
Our discussion thus far has centered around very clean datasets, in which a perfect decision boundary exists. But what if your data has some amount of overlap? For example, you may have data like this:



To deal with data that isn't perfectly separable, SVMs have a built-in flexibility: they "soften" the margin. This means they allow some data points to fall within the margin if it leads to a better overall fit. The "softness" of this margin is controlled by a tuning parameter called C (or the regularization parameter).

The plot below visually demonstrates how changing the value of C affects the final model by adjusting the margin's flexibility.



For very large C, the margin is hard, and points cannot lay in it. For smaller C the margin is softer, and can grow to encompass some points.

Note:
For multiclass classification SVM makes use of `one vs rest, calculating different SVM for each class.

Advantages
Their dependence on relatively few support vectors means that they are very compact models, and take up very little memory.

Once the model is trained, the prediction phase is very fast.

Because they are affected only by points near the margin, they work well with high-dimensional dataeven data with more dimensions than samples, which is a challenging regime for other algorithms.

Their integration with kernel methods makes them very versatile, they are able to adapt to many types of data.

Not very sensitive to overfitting (we will talk about it in detail, in the coming chapters).

Disadvantages
Training time is high when we have large data sets.

When the data set has more noise (i.e. target classes are overlapping) SVM doesnt perform well.
